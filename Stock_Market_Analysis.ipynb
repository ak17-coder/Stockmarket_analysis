```python
# Import libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Download stock data (example: Apple)
stock = yf.download('AAPL', start='2020-01-01', end='2024-12-31')
stock.head()
```

```python
# Plot Closing Price
plt.figure(figsize=(14, 6))
plt.plot(stock['Close'], label='Closing Price')
plt.title('AAPL Closing Price')
plt.xlabel('Date')
plt.ylabel('Price in USD')
plt.legend()
plt.grid()
plt.show()
```

```python
# Calculate moving averages
stock['MA20'] = stock['Close'].rolling(window=20).mean()
stock['MA50'] = stock['Close'].rolling(window=50).mean()
```

```python
# Plot Closing Price with Moving Averages
plt.figure(figsize=(14, 6))
plt.plot(stock['Close'], label='Closing Price')
plt.plot(stock['MA20'], label='20-Day MA')
plt.plot(stock['MA50'], label='50-Day MA')
plt.title('AAPL Stock Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price in USD')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(14, 4))
plt.plot(stock['Volume'], color='gray')
plt.title('AAPL Trading Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.grid()
plt.show()
```
